```python
dataset = pd.read_csv('/kaggle/input/titanic/train.csv')
dataset.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder


X_train = dataset.iloc[:, [2, 4, 5, 6, 7, 9, 11]]
print(X_train.isna().sum())

X_train['Embarked'].fillna(value=X_train['Embarked'].mode()[0], inplace=True)
X_train['Age'].fillna(value=X_train['Age'].mean(), inplace=True)
print(X_train.isna().sum())
X_train = X_train.values

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder="passthrough")
X_train = np.array(ct.fit_transform(X_train))
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [-1])], remainder="passthrough")
X_train = np.array(ct.fit_transform(X_train))

# for i in range(10):
#     print(X[i])

y_train = dataset.iloc[:, 1].values
print(y_train)


# df = pd.DataFrame(dataset, columns=['Fare', 'Pclass'])

# corrMatrix = df.corr()
# print (corrMatrix)
```

```python
testset = pd.read_csv('/kaggle/input/titanic/test.csv')
testset.head()
```

```python
X_test = testset.iloc[:, [1, 3, 4, 5, 6, 8, 10]]
print(X_test.isna().sum())

# plt.hist(X_test['Fare'])

X_test['Fare'].fillna(value=X_test['Fare'].mode()[0], inplace=True)
X_test['Age'].fillna(value=X_test['Age'].mean(), inplace=True)
print(X_test.isna().sum())


X_test = X_test.values

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder="passthrough")
X_test = np.array(ct.fit_transform(X_test))

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [-1])], remainder="passthrough")
X_test = np.array(ct.fit_transform(X_test))


for i in range(10):
    print(X_test[i])
    
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 5:] = sc.fit_transform(X_train[:, 5:])
X_test[:, 5:] = sc.transform(X_test[:, 5:])
```

```python
print(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 5, criterion="entropy", random_state=0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
submit = pd.DataFrame({'PassengerId': testset.iloc[:, 0].values, 'Survived': y_pred})
print(submit)
```

```python
submit.to_csv('titanic.csv', index=False)
```
